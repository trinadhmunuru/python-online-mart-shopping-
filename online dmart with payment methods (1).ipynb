```python
import re  # For basic validation

class Product:
    def __init__(self, pid, name, price, stock):
        self.pid = pid
        self.name = name
        self.price = price
        self.stock = stock

    def __str__(self):
        return f"{self.pid}: {self.name} - ‚Çπ{self.price} ({self.stock} in stock)"


class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password
        self.cart = []

    def add_to_cart(self, product, quantity):
        if product.stock >= quantity:
            self.cart.append((product, quantity))
            product.stock -= quantity
            print(f"‚úÖ Added {quantity} x {product.name} to cart.")
        else:
            print("‚ùå Not enough stock.")

    def view_cart(self):
        print(f"\nüß∫ {self.username}'s Cart:")
        total = 0
        for item, qty in self.cart:
            print(f"{item.name} - ‚Çπ{item.price} x {qty} = ‚Çπ{item.price * qty}")
            total += item.price * qty
        print(f"üí∞ Total: ‚Çπ{total}")
        return total

    def checkout(self, total):
        if total == 0:
            print("üõí Your cart is empty.")
            return False
        print("üßæ Proceeding to payment...")
        payment = Payment()
        if payment.process_payment(total):
            print("‚úÖ Payment Successful! Thank you for shopping at D-Mart.")
            self.cart = []  # Clear cart after successful payment
            return True
        else:
            print("‚ùå Payment Failed. Try again.")
            return False


class Store:
    def __init__(self):
        self.products = []
        self.load_products()

    def load_products(self):
        self.products.append(Product(1, "Rice 5kg", 250, 20))
        self.products.append(Product(2, "Oil 1L", 150, 50))
        self.products.append(Product(3, "Toothpaste", 40, 100))
        self.products.append(Product(4, "Soap", 30, 60))

    def show_products(self):
        print("\nüõí Products Available:")
        for p in self.products:
            print(p)

    def get_product_by_id(self, pid):
        for p in self.products:
            if p.pid == pid:
                return p
        return None


class Payment:
    def process_payment(self, amount):
        print("\nüí≥ Choose Payment Method:")
        print("1. UPI")
        print("2. Credit/Debit Card")
        method = input("Enter choice (1/2): ")

        if method == '1':
            return self.upi_payment(amount)
        elif method == '2':
            return self.card_payment(amount)
        else:
            print("‚ùå Invalid payment method")
            return False

    def upi_payment(self, amount):
        upi = input("Enter your UPI ID (e.g. name@bank): ")
        if re.match(r'^\w+@\w+$', upi):
            print(f"üîÅ Processing ‚Çπ{amount} via UPI...")
            return True
        else:
            print("‚ùå Invalid UPI format.")
            return False

    def card_payment(self, amount):
        card = input("Enter 16-digit card number: ")
        cvv = input("Enter 3-digit CVV: ")
        if card.isdigit() and len(card) == 16 and cvv.isdigit() and len(cvv) == 3:
            print(f"üîÅ Charging ‚Çπ{amount} to your card...")
            return True
        else:
            print("‚ùå Invalid card details.")
            return False


class OnlineDmart:
    def __init__(self):
        self.users = []
        self.store = Store()
        self.current_user = None

    def register(self):
        uname = input("Choose username: ")
        pwd = input("Choose password: ")
        self.users.append(User(uname, pwd))
        print("‚úÖ Registered successfully.")

    def login(self):
        uname = input("Username: ")
        pwd = input("Password: ")
        for u in self.users:
            if u.username == uname and u.password == pwd:
                self.current_user = u
                print(f"üëã Welcome, {u.username}!")
                return
        print("‚ùå Invalid credentials.")

    def shopping_menu(self):
        while True:
            print("\nüìã Shopping Menu")
            print("1. View Products")
            print("2. Add to Cart")
            print("3. View Cart")
            print("4. Checkout")
            print("5. Logout")
            choice = input("Choose an option: ")

            if choice == '1':
                self.store.show_products()
            elif choice == '2':
                try:
                    pid = int(input("Enter product ID: "))
                    qty = int(input("Enter quantity: "))
                    product = self.store.get_product_by_id(pid)
                    if product:
                        self.current_user.add_to_cart(product, qty)
                    else:
                        print("‚ùå Product not found.")
                except ValueError:
                    print("‚ùå Invalid input.")
            elif choice == '3':
                self.current_user.view_cart()
            elif choice == '4':
                total = self.current_user.view_cart()
                self.current_user.checkout(total)
            elif choice == '5':
                print("üëã Logging out...")
                self.current_user = None
                break
            else:
                print("‚ùå Invalid option")

    def run(self):
        while True:
            print("\nüîêtrk Main Menu")
            print("1. Register")
            print("2. Login")
            print("3. Exit")
            ch = input("Enter choice: ")
            if ch == '1':
                self.register()
            elif ch == '2':
                self.login()
                if self.current_user:
                    self.shopping_menu()
            elif ch == '3':
                print("üö™ Exiting... Thank you for visiting Online D-Mart.")
                break
            else:
                print("‚ùå Invalid choice")


# Run the app
app = OnlineDmart()
app.run()
```
